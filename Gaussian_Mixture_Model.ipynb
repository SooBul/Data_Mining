#**Gaussian Mixture Model**

```python
from sklearn.mixture import GaussianMixture
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from numpy import random
from pandas import DataFrame
from sklearn.cluster import KMeans

random.seed (234)
x, _ = make_blobs (n_samples=330, centers=5, cluster_std=1.84)
plt.figure(figsize=(8, 6))
plt.scatter(x[:,0], x[:,1])
plt.show()
```

```python
gm = GaussianMixture (n_components=5).fit(x)
centers=gm.means_
print(centers)


```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x[:,0], x[:,1], label="Các điểm dữ liệu")
plt.scatter(centers[:,0], centers[:,1],c='r', label="Các tâm của cụm")
plt.legend()
plt.show()
```

```python
pred=gm.predict(x)
df = DataFrame({'x':x[:,0], 'y':x[:,1], 'label': pred})
groups= df.groupby('label')

ig, ax = plt.subplots (figsize=(8, 6))
for name, group in groups:
    ax.scatter (group.x, group.y, label=name)
ax.legend()
plt.show()
```

```python
f = plt.figure (figsize=(8, 6), dpi=80)
f.add_subplot(2, 2, 1)
for i in range (2, 6):
    gm= GaussianMixture (n_components=i).fit(x)
    pred=gm.predict(x)
    df = DataFrame({'x':x[:,0], 'y':x[:,1], 'label':pred})
    groups=df.groupby('label')

    f.add_subplot (2, 2, i-1)
    for name, group in groups:
     plt.scatter(group.x, group.y, label=name, s=8)
     plt.title(" (GMM) Cluster size:" + str(i))
     plt.legend ()
plt.tight_layout ()
plt.show()
```

```python
f = plt.figure(figsize=(8, 6), dpi=80)
f.add_subplot (2, 2, 1)
for i in range (2, 6):
 km = KMeans(n_clusters=i). fit (x)
 pred = km.predict(x)
 df = DataFrame ({'x':x[:,0], 'y':x[:,1], 'label': pred})
 groups= df.groupby('label')
 f.add_subplot(2, 2, i-1)
 for name, group in groups:
     plt.scatter (group.x, group.y, label=name, s=8)
     plt.title(" (KMeans) Cluster size:" + str(i))
     plt.legend ()
plt.tight_layout()
plt.show()
```

#**Thuật toán Mahalanobis k-means**





```python
import random
import numpy as np
from scipy.spatial import distance

class Mahalanobis_kmeans:
  def __init__(self, k, initial_iterations=5, iterations=100):
        self.k = k
        self.initial_iterations = initial_iterations
        self.iterations = iterations
        self.seed = 100

  def mahalanobis_distance(self, u, v, cov_inv):
        return np.sqrt(np.dot(np.dot(u - v, cov_inv), (u - v)))

  def fit(self, dataset):

        random.seed(self.seed)

        self.centroids = list(random.sample(list(dataset), 1))

        for i in range(self.k - 1):
            dist = np.array([min([np.linalg.norm(sample - centroid)**2 for centroid in self.centroids]) for sample in dataset])
            index = np.random.choice(range(len(dataset)), p=(dist/dist.sum()))
            self.centroids.append(dataset[index])

        for i in range(self.initial_iterations):

            self.partitions = {i:[] for i in range(self.k)}

            for sample in dataset:
                distances = [np.sqrt(distance.euclidean(sample, centroid)) for centroid in self.centroids]
                self.partitions[distances.index(min(distances))].append(sample)


            for cluster in self.partitions:
                self.centroids[cluster] = np.mean(self.partitions[cluster], axis=0)


        for i in range(self.iterations):

            old_partitions = dict(self.partitions)

            self.partitions = {i:[] for i in range(self.k)}


            for sample in dataset:
                distances = [self.mahalanobis_distance(np.mean(cluster, axis=0), sample, np.linalg.inv(np.cov(cluster, rowvar=False))) for cluster in old_partitions.values()]
                self.partitions[distances.index(min(distances))].append(sample)


            clusters_comparison = [np.allclose(np.mean(self.partitions[cluster]), np.mean(old_partitions[cluster])) for cluster in self.partitions]

            if False not in clusters_comparison:
                return self
                break

        return self

  def predict_Mahalanobis_kmeans(self, dataset):
        self.labels = []
        for sample in dataset:
            distances = [self.mahalanobis_distance(np.mean(cluster, axis=0), sample, np.linalg.inv(np.cov(cluster, rowvar=False))) for cluster in self.partitions.values()]
            self.labels.append(distances.index(min(distances)))
        return self.labels
```

```python

f = plt.figure(figsize=(8, 6), dpi=80)
f.add_subplot (2, 2, 1)
for i in range (2, 6):
 km = Mahalanobis_kmeans(k=i). fit (x)
 pred = km.predict_Mahalanobis_kmeans(x)
 df = DataFrame ({'x':x[:,0], 'y':x[:,1], 'label': pred})
 groups= df.groupby('label')
 f.add_subplot(2, 2, i-1)
 for name, group in groups:
     plt.scatter (group.x, group.y, label=name, s=8)
     plt.title(" (KMeans) Cluster size:" + str(i))
     plt.legend ()
plt.tight_layout()
plt.show()
```

#Nhận xét so sánh các thuật toán phân cụm

Kết quả gom cụm sử dụng K-means và Gaussian Mixture Model cho kết quả giống nhau.

Kết quả gom cụm sử dụng Mahalanobis k-means cho kết quả khác với 2 thuật toán trên.

Kết quả gom cụm sử dụng Mahalanobis k-means có thể khác với kết quả của thuật toán K-means và Gaussian Mixture Model vì các thuật toán này có cách tiếp cận khác nhau để xác định cụm và tính toán khoảng cách.

1. K-means: Thuật toán K-means phân cụm dựa trên việc tìm các centroid và gán các điểm dữ liệu vào cụm gần nhất dựa trên khoảng cách Euclidean. Khi các centroid không thay đổi, thuật toán dừng lại và cho kết quả gom cụm.

2. Gaussian Mixture Model (GMM): GMM là một mô hình xác suất mà mỗi cụm được biểu diễn bằng một phân phối Gaussian. GMM sử dụng kỹ thuật cực đại hợp lý (maximum likelihood) để xác định các tham số của các phân phối Gaussian và dùng chúng để xác suất của mỗi điểm dữ liệu thuộc về từng cụm. Kết quả gom cụm dựa trên các xác suất này.

3. Mahalanobis k-means: Mahalanobis k-means sử dụng khoảng cách Mahalanobis, là một phép đo khoảng cách dựa trên ma trận hiệp phương sai. Phương pháp này giúp xử lý các tập dữ liệu không có phân phối chuẩn hoặc có các đặc trưng tương quan mạnh. Khi tính toán khoảng cách, Mahalanobis k-means sử dụng ma trận nghịch đảo của ma trận hiệp phương sai để chuẩn hóa các trục. Điều này làm cho thuật toán nhạy cảm hơn đối với hình dạng, kích thước và hướng của các cụm.

Do cách tính toán khoảng cách khác nhau, Mahalanobis k-means có thể tạo ra các cụm có hình dạng, kích thước và hướng khác nhau so với K-means và GMM. Mahalanobis k-means có thể hiệu quả hơn trong việc phân cụm dữ liệu có đặc trưng không phân phối chuẩn hoặc có mối tương quan mạnh giữa các biến. Tuy nhiên, việc lựa chọn phương pháp gom cụm phù hợp phụ thuộc vào tính chất của dữ liệu và mục tiêu phân tích.
